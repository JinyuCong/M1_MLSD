Matplotlib est une bibliothèque pour la création de visualisations statiques, animées et interactives en Python.

## Les bases

La manière standard d'importer matplotlib est la suivante:

```python
import matplotlib.pyplot as plt
```

Matplotlib plot vos données sur des Figures (par exemple, des fenêtres, des widgets Jupyter, etc.), chacune pouvant contenir un ou plusieurs Axes, une zone où les points peuvent être spécifiés en termes de coordonnées x-y (ou theta-r dans un tracé polaire, x-y-z dans un tracé 3D, etc.) 

La façon la plus simple de créer une figure avec un axe est d'utiliser pyplot.subplots. Nous pouvons ensuite utiliser Axes.plot pour dessiner des données sur les axes, et show pour afficher la figure :


```python
import numpy as np

fig, ax = plt.subplots()             # Create a figure containing a single Axes.
ax.plot(np.arange(0, np.pi*4, .1), np.sin(np.arange(0, np.pi*4, .1)))  # Plot some data on the Axes.
plt.title('Function Sinus')               # Set the title of the figure.
plt.xlabel('x')                          # Set the x-axis label of the current axes.
plt.ylabel('sin(x)')                     # Set the y-axis label of the current axes.
plt.show()                           # Show the figure.
```

Les composants d'une figure sont les suivants :

<img src="https://matplotlib.org/stable/_images/anatomy.png" width="500" >

Matplotlib utilise une logique orientée objet, ce qui signifie que les graphiques sont construits à partir de composants plus petits. Par exemple, il crée des objets :

- Figure : La fenêtre ou le canevas sur lequel les graphiques sont dessinés.
- Axes : Les axes du graphique, où les données sont tracées.
- Axis : Les lignes numérotées et les étiquettes qui délimitent les axes.
- Artist : Tout ce qui est dessiné sur la figure (par exemple, les lignes, les légendes, les titres, etc.).



## Types de graphiques

Matplotlib prend en charge de nombreux types de graphiques différents. Voici quelques-uns des plus courants :

- Line plot : Tracer des points reliés par des lignes.
- Scatter plot : Tracer des points sans les relier.
- Bar plot : Tracer des barres verticales ou horizontales.
- Histogram : Tracer des histogrammes.
- Pie plot : Tracer des diagrammes circulaires.
- Box plot : Tracer des boîtes à moustaches.

Nous allons maintenant voir comment créer ces différents types de graphiques en utilisant Matplotlib. 


```python
# Graphique en ligne
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4, 5, 6], [1, 4, 2, 3, 7, 1])
plt.show()
```

```python
# Graphique en nuage de points

fig, ax = plt.subplots()
ax.scatter([1, 2, 3, 4, 5, 6], [1, 4, 2, 3, 7, 1])
plt.show()
```

```python
# Graphique en barres

fig, ax = plt.subplots()
ax.bar([1, 2, 3, 4, 5, 6], [1, 4, 2, 3, 7, 1])
plt.show()
```

## Personnalisation des graphiques

Matplotlib offre de nombreuses options pour personnaliser vos graphiques. Vous pouvez modifier les couleurs, les styles de ligne, les étiquettes, les titres, les légendes, les limites des axes, etc. Voici quelques exemples de personnalisation :


```python
# Les couleurs 

data = np.random.normal(10, 2, size=(100, 2))

fig, ax = plt.subplots()
ax.scatter(data[:, 0], data[:, 1], color='red')
plt.show()
```

```python
# Les styles de ligne

fig, ax = plt.subplots()
ax.plot(np.arange(0, np.pi*4, .1), np.sin(np.arange(0, np.pi*4, .1)), linestyle='--')
plt.show()
```

```python
# Les étiquettes

fig, ax = plt.subplots()
ax.plot(np.arange(0, np.pi*4, .1), np.sin(np.arange(0, np.pi*4, .1)))
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Function Sinus')
plt.show()
```

```python
# Les marqueurs et légendes
data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets

fig, ax = plt.subplots()

ax.plot(data1, 'o', label='data1')
ax.plot(data2, 'd', label='data2')
ax.plot(data3, 'v', label='data3')
ax.plot(data4, 's', label='data4')

plt.legend()
plt.show()



```

```python
# Les limites des axes

fig, ax = plt.subplots()
ax.plot(np.arange(0, np.pi*4, .1), np.sin(np.arange(0, np.pi*4, .1)))
ax.set_xlim([0, 10])
ax.set_ylim([-1, 1])

plt.show()
```

```python
# Les axis

fig, ax = plt.subplots()
ax.plot(np.arange(0, np.pi*4, .1), np.sin(np.arange(0, np.pi*4, .1)))
ax.set_xlim([0, np.pi*4])
ax.set_ylim([-1.1, 1.1])
ax.set_xticks([0, np.pi, np.pi*2, np.pi*3, np.pi*4])
ax.set_yticks([-1, 0, 1])
ax.set_xticklabels(['0', 'π', '2π', '3π', '4π'])

plt.show()
```

## Travailler avec plusieurs axes

Vous pouvez également créer des figures avec plusieurs axes. Par exemple, vous pouvez créer une figure avec 4 axes disposés en 2x2 :


```python
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)
axs[0, 0].plot([1, 2, 3, 4], [10, 20, 25, 30])
axs[0, 1].scatter([1, 2, 3, 4], [10, 20, 25, 30])
axs[1, 0].bar([1, 2, 3, 4], [10, 20, 25, 30])
x = np.arange(1, 4, .1)
axs[1, 1].fill_between(x, 3*x, np.exp(.9 * x))

plt.tight_layout()
plt.show()
```

#### Créer des colonnes à partir de données existantes

On peut créer des colonnes calculées à partir de données existantes. Par exemple, pour calculer l'âge dans 10 ans :

```python
data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
        'age': [25, 30, 35, 40],
        'city': ['Paris', 'London', 'New York', 'Berlin']}
df = pd.DataFrame(data)


df['age_dans_10_ans'] = df['age'] + 10

df
```

#### Fusion de DataFrames

Pour la fusion de deux DataFrames. Comme en NumPy, on peut utiliser les fonctions de concatenations :

```python
data1 = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
        'age': [25, 30, 35, 40],
        'city': ['Paris', 'London', 'New York', 'Berlin']}

data2 = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
        'job': ['Engineer', 'Doctor', 'Teacher', 'Pilot']}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df = pd.concat([df1, df2], axis=1)

df
```

Cet opération a introduit une colonne double name. On peut supprimer cette colonne en utilisant la méthode drop :

```python
df.drop('name', axis=1, inplace=True)
```

On voit que les deux colonnes name ont été supprimées. Donc, plutôt que de concaténer les deux DataFrames, on aurait dû les fusionner en utilisant la méthode merge sur la colonne name comme en SQL :



```python
df = pd.merge(df1, df2, on='name')

df
```

C'est l'équivalent d'un INNER JOIN en SQL. Bien sûr, il y a plusieurs manières de faire, pas que celle-ci. 

La fonction merge utilise par défaut une jointure interne. On peut spécifier le type de jointure en utilisant le paramètre 'how'. Les jointures possibles sont : 'left', 'right', 'outer', 'inner'.

#### Grouper les données

On peut grouper les données en utilisant la méthode groupby. Par exemple, pour calculer la moyenne de l'âge par ville :

```python
data = {'name': ['Alice', 'Bob', 'Charlie', 'David', 'Alice', 'Bob', 'Charlie', 'David'],
        'age': [25, 30, 35, 40, 25, 30, 35, 40],
        'city': ['Paris', 'London', 'New York', 'Berlin', 'Paris', 'Berlin', 'New York', 'London']}
df = pd.DataFrame(data)

df.groupby('city').mean('age')
```

Chaque ligne de ce tableau donne des informations sur le nom d'un pays, le continent auquel il appartient, sa superficie, sa population et la valeur de son PIB.

Un pays est grand si

- il a une superficie d'au moins trois millions de kilomètres (soit 3000000 km2), ou
- il a une population d'au moins vingt-cinq millions d'habitants (soit 25000000).

Rédigez une solution pour trouver le nom, la population et la superficie des grands pays. Retournez le tableau des résultats dans n'importe quel ordre.

Le format du résultat est celui de l'exemple suivant :

Example 1:
<pre>
Input: 
World table:
+-------------+-----------+---------+------------+--------------+
| name        | continent | area    | population | gdp          |
+-------------+-----------+---------+------------+--------------+
| Afghanistan | Asia      | 652230  | 25500100   | 20343000000  |
| Albania     | Europe    | 28748   | 2831741    | 12960000000  |
| Algeria     | Africa    | 2381741 | 37100000   | 188681000000 |
| Andorra     | Europe    | 468     | 78115      | 3712000000   |
| Angola      | Africa    | 1246700 | 20609294   | 100990000000 |
+-------------+-----------+---------+------------+--------------+

Output: 
+-------------+------------+---------+
| name        | population | area    |
+-------------+------------+---------+
| Afghanistan | 25500100   | 652230  |
| Algeria     | 37100000   | 2381741 |
+-------------+------------+---------+




```python
def find_big_countries(world):
    df = world[(world['area'] > 3000000) | (world['population'] >= 25000000)]
    df = df[]
    print(df)
```

```python
# Test the function with the provided sample input

world_data = {
    'name': ['Afghanistan', 'Albania', 'Algeria', 'Andorra', 'Angola'],
    'continent': ['Asia', 'Europe', 'Africa', 'Europe', 'Africa'],
    'area': [652230, 28748, 2381741, 468, 1246700],
    'population': [25500100, 2831741, 37100000, 78115, 20609294],
    'gdp': [20343000000, 12960000000, 188681000000, 3712000000, 100990000000]
}

world_df = pd.DataFrame(world_data)

results_data = {
    'name': ['Afghanistan', 'Algeria'],
    'population': [25500100, 37100000],
    'area': [652230, 2381741]
}

results_df = pd.DataFrame(results_data)

user_result = find_big_countries(world_df)

assert (results_df.sort_values(by='name') == user_result.sort_values(by='name')).all(axis=None)
assert (results_df.columns == user_result.columns).all()
```
