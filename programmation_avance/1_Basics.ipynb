# <span style='color:green; font-weight: bold;'> Pourquoi Python ? </span>

### **Simple mais Puissant**
- Python est connu pour sa syntaxe claire et lisible. Il utilise l'indentation pour délimiter les blocs de code, ce qui rend le code plus facile à lire et à comprendre
- Python est un langage interprété, ce qui signifie que le code est exécuté ligne par ligne. Cela permet une grande interactivité et facilite le débogage et la portabilité.
- Python est un langage à typage dynamique, ce qui signifie que les types de variables sont déterminés au moment de l'exécution et non à la compilation. Cela permet une plus grande flexibilité dans le code.
- Python est un langage fortement typé, il ne fait pas de conversion implicite entre les types.
- Python supporte plusieurs paradigmes de programmation, y compris la programmation :
  - procédurale (scripts et fonctions),
  - orientée objet (classes, héritage, encapsulation),
  - fonctionnelle (fonctions anonymes, map, filter, reduce, etc.).

## Compréhensions
Un raccourci pratique pour définir une liste, un set ou un dictionnaire est la compréhension.

Elle offre une manière compacte de le faire. Par exemple si nous avons une liste de nombres et qu'on veut créer une deuxième qui contient les carrés de ces nombres, pour la définir, on peut écrire:

````python
x = [1, 2, 3, 4, 5]

x2 = []
for e in x:
    x2.append(e**2)

x2
````

On pourra l'écrire en une ligne en utilisant les compréhensions

````python
x2 = [e**2 for e in x]

x2
````

Il y a moyen d'utiliser les comprehensions pour créer différentes collections et pas seulement les lists. Voici le moyen de le faire pour une liste, dictionnaire et set.  



````python
n = 10

list_comp = [x**2 for x in range(n)]
dict_comp = {f'{x}**2':x**2 for x in range(n)}
set_comp = {x**2 for x in range(n)}

print(list_comp)
print(dict_comp)
print(set_comp)
````

````python
print(type(list_comp))
print(type(dict_comp))
print(type(set_comp))
````

## Les boucles

Il existe deux types de boucles en Python, la boucle ```for``` et la boucle ```while```.

### La boucle ```for```

La boucle ```for``` est utilisée pour répéter un bloc de code un nombre spécifié de fois. On peut utiliser la fonction ```range``` pour générer une séquence de nombres à utiliser dans la boucle ```for```. Par exemple, pour afficher les nombres de 0 à 2, on peut écrire :

````python
for i in range(3):
    print(i)
````

La syntaxe de la fonction ```range``` est ```range(start, stop, step)```. Par défaut, ```start``` est 0 et ```step``` est 1. Par exemple, pour afficher les nombres de 5 à 7, on peut écrire :

````python
for i in range(5, 8):
    print(i)
````

````python
for i in range(4, 9, 2):
    print(i)
````

````python
for i in range(8, 3, -2):
    print(i)
````

Pour parcourir une liste, on peut utiliser la boucle ```for``` de la manière suivante:

````python
x = [1, 2, 3, 4, 5]

for i in range(len(x)):
    print('L\'element', i, 'est égale à', x[i])
````

### Coding problem 1 (lists)

Étant donné deux tableaux ```nums1``` et ```nums2``` triés en ordre croissant, fusionnez-les en un seul tableau trié en ordre croissant.

Contraintes:
- Vous devez résoudre le problème sans utiliser de fonction de tri.
- Vous ne pouvez pas utiliser de tableaux supplémentaires. Vous devez modifier le tableau ```nums1``` en place.
- Vous ne pouvez faire qu'un seul passage sur ```nums1``` et ```nums2```.
- Ne pas mettre de return dans la fonction.

````python
def merge(nums1, nums2):
    # write your code here
    # i=0
    # nums1
    # [1,2,6]
    # [3,5,7]
    # 
    # [1,2,3,5,6,7]
    
    i = 0
    j = 0
    res = []
    while i < len(nums1) and j < len(nums2):
        if nums1[i] < nums2[j]:
            res.append(nums1[i])
            i += 1
        else:
            res.append(nums2[j])
            j += 1
        
    if i < len(nums1):
        res.extend(nums1[i:])
    if j < len(nums2):
        res.extend(nums2[j:])
        
    return res
    
a = [1,2,6,8,10,11,12,13]
b = [3,5,7]
merge(a, b)
````

### Coding problem 2 (lists+sets)

Étant donné un tableau d'entiers 2D ```nums``` où ```nums[i]``` est un tableau non vide d'entiers positifs distincts. Renvoyer la liste des entiers qui sont présents dans tous les tableaux présents dans ```nums``` triés par ordre décroissant.

Contraintes:
- Toutes les valeurs de ```nums[i]``` sont uniques.

Exemple:

- entrées:
    ```nums = [[1,2,3,4], [2,3,4,5], [3,4,5,6]]```
- sortie:
    ```[4, 3]```

````python
def intersection(nums):
    # write your code here
    inter = set(nums[0])
    for i in range(1, len(nums)):
        inter = inter.intersection(set(nums[i]))
    
    return sorted(list(inter), reverse=True)
    
nums = [[1,2,3,4], [2,3,4,5], [3,4,5,6]]
intersection(nums)
````

### Coding problem 3 (lists+dictionnaries)

Étant donné un tableau d'entiers ```nums``` et un entier ```target```, renvoyer les indices des deux nombres tels que leur somme soit égale à target.
Vous ne pouvez pas utiliser deux fois le même élément.

Contraintes:
- Commencer par implémenter en *Brute Force* par la recherche exhaustive.
- Ensuite, implémenter une solution plus efficace en utilisant les dictionnaires.
- On suppose qu'une solution existe toujours.

````python
def brute_force_solution(nums, target):
    # write your code here
    for i in range(len(nums)):
        for j in range(len(nums)):
            if nums[i] + nums[j] == target:
                return i, j

def efficient_solution(nums, target):
    # write your code here
    dic = {num: ids for ids, num in enumerate(nums)}
    for num in dic:
        to_search = target - num
        if to_search in dic:
            return dic[num], dic[to_search]

nums = [1,3,4,5,6,12]

efficient_solution(nums, 11)
````

### Coding problem 5 (coder une régression linéaire simple)

Nous souhaitons coder une régression linéaire simple c'est à dire avec une seule variable explicative $\mathbf{x}$ et une seule variable réponse $\mathbf{y}$. L'equation d'une droite est donné comme:
$$\mathbf{y}=\alpha+\beta\mathbf{x}$$

Comme la relation entre une variable réponse $\mathbf{y}$ et une variable explicative $\mathbf{x}$ n'est pas forcément linéaire le modèle linéaire simple introduit la notion d'erreur. Nous pouvons décrire la relation sous-jacente entre $y_i$ et $x_i$ impliquant ce terme d'erreur $\varepsilon_i$ par
$$y_i = \alpha + \beta x_i + \varepsilon_i$$

Le but est de trouver les estimations de $\alpha$ et $\beta$ qu'on notera $\widehat{\alpha}$ et $\widehat{\beta}$ en minimisant l'erreur estimée pour n'impore quel couple $\alpha, \beta$:
$${\widehat {\varepsilon }}_{i}=y_{i}-\alpha -\beta x_{i}$$

Il en découle le problème de minimisation suivant :

$$\min_{\alpha,\beta} \frac{1}{n}\sum_{i=1}^n \varepsilon_i^2 = \frac{1}{n}\sum_{i=1}^n (y_{i}-\alpha -\beta x_{i})^2$$

Il existe deux manière de résoudre ce problème d'optimisation:
1. Analytiquement en trouvant les solutions optimales directement.
2. Numériquement en utilisant des algorithmes d'optimisation comme la descente de gradient.


#### Solution Analytique optimale

On commence par la solution exacte. En prenant la dérivée par rapport à $\alpha$ et $\beta$, on obtient les solutions optimales qui sont données par :
$${\beta_{optimal}}={\frac {\sum _{i=1}^{n}(x_{i}-{\bar {x}})(y_{i}-{\bar {y}})}{\sum _{i=1}^{n}(x_{i}-{\bar {x}})^{2}}}$$

$${\textstyle {\begin{aligned}{\ \alpha_{optimal} }&={\bar {y}}-{ {\beta }_{optimal}}\,{\bar {x}}\end{aligned}}}$$

Le but est que vous implémentiez ces solutions en Python.

````python
def linear_regression(x, y):
    # write your code here
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    cov = np.dot((x - mean_x), (y - mean_y))
    var_x = np.sum((x - mean_x)**2)
    beta = cov / var_x
    alpha = mean_y - beta * mean_x
    
    return beta, alpha
````

````python
import numpy as np
import matplotlib.pyplot as plt


def generate_test_case(max_len=100):
    np.random.seed(0)
    x = np.arange(0, max_len)
    a = np.random.rand()+1
    noise = np.random.normal(2, 5, len(x))
    y = a * x + noise + 20
    y = y / 10
    return x, y

beta, alpha = linear_regression(x, y)
plt.scatter(x, y, alpha=.66)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,
         [beta*xi+alpha for xi in x],
         '--r',
         label=f'y={beta:.2f}x+{alpha:.2f}')
plt.legend()
plt.show()
````

#### Solutions par Descente de Gradient

Une manière itérative d'avoir les solutions est d'utiliser la descente de gradient. On considère la même fonction objectif

$$J(\alpha, \beta)=\sum_{i=0}^n (y_{i}-(\alpha +\beta x_{i}))^2$$

On peut optimiser cette fonction petit à petit en faisant des mises à jour comme suit :

$$\alpha=\alpha - r_1\frac{\partial J}{\partial \alpha}$$

$$\beta=\beta - r_2\frac{\partial J}{\partial \beta}$$

où $r_1$ ($r_2$) est un paramètre appelé learning rate et les gradients sont donnés par :

$$\frac{\partial J}{\partial \beta}=\frac{-2}{n}\sum_{i=0}^n x_i(y_i-(\beta x_i+\alpha))$$
$$\frac{\partial J}{\partial \alpha}=\frac{-2}{n}\sum_{i=0}^n (y_i-(\beta x_i+\alpha))$$

Le but est que vous implémentiez ces règles de mise à jour ainsi que la fonction objectif en Python.




````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython import display

def grad_alpha(x, y, alpha, beta):
    # write your code here
    n = len(x)
    return -2/n * np.sum(y - (beta * x + alpha))

def grad_beta(x, y, alpha, beta):
    # write your code here
    n = len(x)
    return -2/n * np.sum(x * (y - (beta * x + alpha)))

def loss_fn(x, y, alpha, beta):
    # write your code here
    return np.sum((y - (beta * x + alpha))**2)

def linear_regression_gradient_descent_update(x, y, alpha, beta, r1=2e-2, r2=1e-4):
    # write your code here
    prev_alpha = alpha
    prev_beta = beta
    alpha = alpha - r1 * grad_alpha(x, y, prev_alpha, prev_beta)
    beta = beta - r2 * grad_beta(x, y, prev_alpha, prev_beta)
    return alpha, beta
````

````python
def generate_test_case(max_len=100):
    np.random.seed(0)
    x = np.arange(0, max_len)
    a = np.random.rand()+1
    noise = np.random.normal(2, 5, len(x))
    y = a * x + noise + 2
    y = y / 10
    return x, y


x, y = generate_test_case()


alpha, beta = 10, 3

alphas = [alpha]
betas = [beta]

old_loss = 10e9
loss = loss_fn(x, y, alpha, beta)

i = 0

while abs(loss - old_loss) > 10e-5 and i < 1e3:
    alpha, beta = linear_regression_gradient_descent_update(x, y, alpha, beta)
    alphas.append(alpha)
    betas.append(beta)
    old_loss = loss
    loss = loss_fn(x, y, alpha, beta)
    i += 1

print(beta, alpha)

fig, ax = plt.subplots()
ax.scatter(x, y, alpha=.66)
ax.set_xlabel('x')
ax.set_ylabel('y')
line, = ax.plot([], [], '--r',
             label=f'y={beta:.2f}x+{alpha:.2f}')
line.set_data([x, [betas[i]*xi+alphas[i] for xi in x]])
def animate(i):
    line.set_data([x, [betas[i]*xi+alphas[i] for xi in x]])
    ax.set_title(f'Fit at iteration {i+1}')
    line.set_label(f'y={betas[i]:.2f}x+{alphas[i]:.2f}')
    ax.legend(loc='lower right')

anim = FuncAnimation(fig, animate, frames=len(alphas), interval=600)

video = anim.to_html5_video()
html = display.HTML(video)
plt.close()

display.display(html)
````
