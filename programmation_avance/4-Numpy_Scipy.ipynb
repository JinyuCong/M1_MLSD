# <span style='color:green; font-weight: bold;'> L'algèbre Linéaire avec NumPy </span>

L'algèbre linéaire est une branche des mathématiques qui étudie les vecteurs, les espaces vectoriels, les transformations linéaires, les systèmes d'équations linéaires, les valeurs propres, entre autres. Les vecteurs et les matrices sont des structures de données fondamentales pour l'algèbre linéaire. Les vecteurs sont des tableaux unidimensionnels tandis que les matrices sont des tableaux bidimensionnels.

NumPy fournit des fonctions pour effectuer des opérations d'algèbre linéaire telles que la résolution de systèmes d'équations linéaires, le calcul de valeurs propres, le calcul de la norme d'un vecteur, le produit scalaire de vecteurs, le produit matriciel, etc.

# Les produits matriciels et vectoriels

Pour faire un produit matriciel, on utilise la fonction `dot` de NumPy. Par exemple, pour multiplier deux matrices `A` et `B`, on utilise `np.dot(A, B)`. Si `A` et `B` sont des tableaux bidimensionnels, `np.dot(A, B)` effectue un produit matriciel. Si `A` et `B` sont des tableaux unidimensionnels, `np.dot(A, B)` effectue un produit scalaire.








```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(A @ B)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print('Produit scalaire:', a @ b) # (3,) @ (3,) = (1,)
print('Inner product:', np.inner(a, b)) # (3,) @ (3,) = (1,)

print('---------------------------------')

print('Produit tensoriel:', a[:, np.newaxis] @ b[np.newaxis]) # (3, 1) @ (1, 3) = (3, 3)
print('Outer product:', np.outer(a, b))
```

La puissance d'une matrice est une opération qui consiste à multiplier une matrice par elle-même un certain nombre de fois en terme de produits matriciels. Par exemple, pour calculer la puissance 3 d'une matrice `A`, on utilise `np.linalg.matrix_power(A, 3)`.

```python
A = np.array([[1, 2], [3, 4]])

print(A @ A @ A)

print(np.linalg.matrix_power(A, 3))
```

# Décompositions de matrices

 Il existe plusieurs décompositions de matrices en algèbre linéaire. Les décompositions les plus courantes sont la décomposition en valeurs singulières (SVD), la décomposition en valeurs propres et vecteurs propres, la décomposition QR, la décomposition LU, etc.
 
Ces décompositions permettent de résoudre certains problèmes d'optimisation, de réduire la dimensionnalité des données, de compresser des images, etc. 

Par exemple pour la SVD décompose une matrice `A` en trois matrices `U`, `S` et `V` telles que `A = U @ S @ V`. La matrice `U` est une matrice orthogonale, `S` est une matrice diagonale et `V` est une matrice orthogonale :

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/b/bb/Singular-Value-Decomposition.svg/1920px-Singular-Value-Decomposition.svg.png" width="400">
 
NumPy fournit des fonctions pour effectuer ces décompositions. Par exemple :
 
| Fonction             | Description                                                                 |
|----------------------|-----------------------------------------------------------------------------|
| `np.linalg.svd`      | Décomposition en valeurs singulières (existe pour n'importe quelle matrice) |
| `np.linalg.eig`      | Décomposition en valeurs propres (existe pour les matrices carrées)         |                                
| `np.linalg.qr`       | Décomposition QR                                                            |
| `np.linalg.cholesky` | Décomposition de Cholesky                                                   |

 

```python
A = np.array([[1, 2], [3, 4]])

U, S, V = np.linalg.svd(A)
print('Décomposition en valeurs singulières:')
print('U:', U)
print('S:', S)
print('V:', V)

print('---------------------------------')

W, V = np.linalg.eig(A)
print('Décomposition en valeurs propres:')
print('W:', W)
print('V:', V)
```

# Normes et autres nombres spéciaux

Une norme est une fonction qui attribue une longueur ou une taille à un vecteur ou une matrice. Il existe plusieurs normes en algèbre linéaire. Les normes les plus courantes sont la norme L1, la norme L2, la norme infinie, la norme de Frobenius, etc.

NumPy fournit des fonctions pour calculer ces normes. Par exemple :

| Fonction          | Description                   |
|-------------------|-------------------------------|
| `np.linalg.norm`  | Norme d'un vecteur ou matrice |
| `np.linalg.det`   | Déterminant d'une matrice     |
| `np.trace`        | Trace d'une matrice           |
| `np.linalg.matrix_rank` | Rang d'une matrice       |


```python
x = np.array([1, 2, 3, 4])

print('Norme L1:', np.linalg.norm(x, ord=1))
print('Norme L2:', np.linalg.norm(x, ord=2))
print('Norme infinie:', np.linalg.norm(x, ord=np.inf))

print('---------------------------------')

A = np.array([[1, 2], [3, 4]])

print('Norme de Frobenius:', np.linalg.norm(A, ord='fro'))
print('Déterminant:', np.linalg.det(A))
print('Trace:', np.trace(A))
print('Rang:', np.linalg.matrix_rank(A))
```

# Résolution de systèmes d'équations linéaires et inversion de matrices


La résolution de systèmes d'équations linéaires est une opération courante en algèbre linéaire. Un système d'équations linéaires est un ensemble d'équations linéaires de la forme `Ax = b` où `A` est une matrice, `x` est un vecteur inconnu et `b` est un vecteur de constantes.

NumPy fournit des fonctions pour résoudre des systèmes d'équations linéaires et inverser des matrices. Par exemple :

| Fonction          | Description                   |
|-------------------|-------------------------------|
| `np.linalg.solve` | Résoudre un système d'équations linéaires |
| `np.linalg.inv`   | Inverser une matrice           |
| `np.linalg.pinv`  | Pseudo-inverse d'une matrice   |


```python
A = np.array([[1, 2], [3, 4]])
b = np.array([5, 6])

x = np.linalg.solve(A, b)
print('Solution du système d\'équations linéaires:', x)

print('---------------------------------')

A_inv = np.linalg.inv(A)
print('Inverse de la matrice A:', A_inv)

print('---------------------------------')

A = np.array([[1, 2, 3], [2, 4, 5]])
A_pinv = np.linalg.pinv(A)
print('Pseudo-inverse de la matrice A:', A_pinv)
```

## Matrices creuses
Une des fonctionnalités les plus intéressantes de `scipy` est le module `scipy.sparse` qui permet de travailler avec des matrices creuses. Les matrices creuses sont des matrices qui contiennent un grand nombre de zéros. Elles sont utilisées pour économiser de l'espace mémoire et améliorer les performances des opérations matricielles. Elle fournit aussi des fonctions d'algèbre linéaire pour les matrices creuses.

<img src="https://matteding.github.io/images/sparse_dense.gif" width="600">

Il existe plusieurs types de matrices creuses dans `scipy.sparse`. Par exemple :

- `scipy.sparse.csr_matrix` : Format creux compressé en rangée<br><br>
    <img src="https://matteding.github.io/images/csr.gif" width="600"><br>
- `scipy.sparse.csc_matrix` : Format creux compressé en colonne<br><br>
    <img src="https://matteding.github.io/images/csc.gif" width="600"><br>
- `scipy.sparse.coo_matrix` : Format en coordonnées<br><br>
    <img src="https://matteding.github.io/images/coo.gif" width="600"><br>
- `scipy.sparse.dia_matrix` : Format en diagonales<br><br>
    <img src="https://matteding.github.io/images/dia.gif" width="600">







```python
# make a sparse matrix with different formats and a numpy array and compare their sizes
import numpy as np
from scipy.sparse import csr_matrix, csc_matrix, coo_matrix

# Create a dense matrix
dense_matrix = np.random.choice(np.array([0, 1]), size=(3000, 1000), p=[0.95, 0.05]) 

# Create a sparse matrix in CSR format
sparse_csr = csr_matrix(dense_matrix)

# Create a sparse matrix in CSC format
sparse_csc = csc_matrix(dense_matrix)

# Create a sparse matrix in COO format
sparse_coo = coo_matrix(dense_matrix)


print('Taille de la matrice dense:', dense_matrix.nbytes // 1024 ** 2, 'MB')
print('Taille de la matrice creuse CSR:', (sparse_csr.data.nbytes + sparse_csr.indptr.nbytes + sparse_csr.indices.nbytes) // 1024 ** 2, 'MB')
print('Taille de la matrice creuse CSC:', (sparse_csc.data.nbytes + sparse_csc.indptr.nbytes + sparse_csc.indices.nbytes) // 1024 ** 2, 'MB')
print('Taille de la matrice creuse COO:', (sparse_coo.data.nbytes + sparse_coo.row.nbytes + sparse_coo.col.nbytes) // 1024 ** 2, 'MB')


```

```python
import time

start = time.time()
_ = dense_matrix @ dense_matrix.T
end = time.time()
print(f'Temps de multiplication de matrices denses: {end - start:.2f} secondes')

start = time.time()
_ = sparse_csr @ sparse_csr.T
end = time.time()
print(f'Temps de multiplication de matrices creuses CSR: {end - start:.2f} secondes')

start = time.time()
_ = sparse_csc @ sparse_csc.T
end = time.time()

print(f'Temps de multiplication de matrices creuses CSC: {end - start:.2f} secondes')

start = time.time()
_ = sparse_coo @ sparse_coo.T
end = time.time()

print(f'Temps de multiplication de matrices creuses COO: {end - start:.2f} secondes')
```

## Exercice 1

Vous participez à un jeu télévisé. Il y a 3 portes :
- Derrière l'une d'elles se trouve une voiture (gagnant)
- Derrière les deux autres, une chèvre (perdant)

**Déroulement :**
1. Vous choisissez une porte (mais elle n'est pas ouverte).
2. L'animateur, qui sait ce qu'il y a derrière les portes, ouvre une des deux autres portes, révélant une chèvre.
3. Il vous propose alors de **changer de porte** ou de **garder votre choix initial**.

**Question :** Faut-il changer de porte pour maximiser ses chances de gagner la voiture ?

On va simuler ce jeu des milliers de fois pour observer la proportion de victoires selon deux stratégies :
- Toujours garder la porte initiale (`stay`)
- Toujours changer de porte (`switch`)

```python
import random
import numpy as np

max_iters = 1000
res_change = []
res_change_pas = []


for _ in range(max_iters):
    # 创建门
    v_position = np.random.randint(0, 3)
    num_porte = [0, 1, 2]
    portes = ['c', 'c', 'c']
    portes[v_position] = 'v'
    
    # 随机选择一个门
    porte_choisi = np.random.randint(0, 3)
    num_porte.remove(porte_choisi)
    
    # 主持人揭示一个门
    if portes[porte_choisi] == 'v':
        porte_releve = random.sample(num_porte, 1)[0]
    elif portes[porte_choisi] == 'c':
        index_v = portes.index('v')
        num_porte.remove(index_v)
        porte_releve = num_porte[0]
    
    # 不变选择的门
    if portes[porte_choisi] == 'v':
        res_change_pas.append(1)
    else:
        res_change_pas.append(0)
        
    # 变选择的门
    a = [0, 1, 2]
    a.remove(porte_choisi)
    a.remove(porte_releve)
    porte_choisi = a[0]
    if portes[porte_choisi] == 'v':
        res_change.append(1)
    else:
        res_change.append(0)
    
print(f"改变门之后的胜率{sum(res_change) / len(res_change)}")
print(f"不改变门之后的胜率{sum(res_change_pas) / len(res_change_pas)}")
    
```


## Exercice 2

Ecrire une classe `LinearRegression` pour implémenter la régression linéaire avec la méthode des moindres carrés. La classe doit avoir les méthodes suivantes :

- `fit(X, y)`: pour entraîner le modèle sur un ensemble de données `X` et `y`.
- `predict(X)`: pour faire des prédictions sur de nouvelles données `X`.
- `fit_predict(X, y)`: pour entraîner le modèle et faire des prédictions en une seule fois.

et les attributs suivants :

- `coef_`: pour les coefficients de la régression.
- `intercept_`: pour l'ordonnée à l'origine.

La formule de la régression linéaire est la suivante :

$$
y = X\beta + \alpha
$$

où `X` est la matrice des données de taille `m x n`, $\beta$ est le vecteur des coefficients de taille `n` et $\alpha$ est l'ordonnée à l'origine.

La solution des moindres carrés est donnée par la formule suivante :

$$
\beta = (X^T X)^{-1} X^T y
$$

et

$$
\alpha = \bar{y} - \bar{X}\beta
$$

Où $\bar{y}$ est la moyenne des valeurs de `y` et $\bar{X}$ est la moyenne des valeurs de `X` sur les lignes.







```python
class LinearRegression:
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None
    
    def fit(self, X, y):
        self.coef_ = np.linalg.inv(X.T @ X) @ X.T @ y
        self.intercept_ = np.mean(y) - np.mean(X, axis=0) @ self.coef_
    
    def predict(self, X):
        return X @ self.coef_ + self.intercept_
    
    def fit_predict(self, X, y):
        self.fit(X, y)
        return self.predict(X)
    
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X_train = np.random.rand(2000).reshape(1000,2)*60
y_train = (X_train[:, 0]**2)+(X_train[:, 1]**2)
X_test = np.random.rand(200).reshape(100,2)*60
y_test = (X_test[:, 0]**2)+(X_test[:, 1]**2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_train[:,0], X_train[:,1], y_train, marker='.', color='red')
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("y")

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("MAE: {}".format(np.abs(y_test-y_pred).mean()))
print("RMSE: {}".format(np.sqrt(((y_test-y_pred)**2).mean())))

coefs = model.coef_
intercept = model.intercept_
xs = np.tile(np.arange(61), (61,1))
ys = np.tile(np.arange(61), (61,1)).T
zs = xs*coefs[0]+ys*coefs[1]+intercept
print("Equation: y = {:.2f} + {:.2f}x1 + {:.2f}x2".format(intercept, coefs[0],
                                                          coefs[1]))

ax.plot_surface(xs,ys,zs, alpha=0.5)
plt.show()
```

Le pseudo-code de la fonction power_iteration est le suivant :

<pre>

2. Normaliser v en le divisant par sa norme ||v||.
   
3. Répéter pour max_iter itérations :

      v_new = A^T @ u
   d. Normaliser v_new en le divisant par sa norme ||v_new||.
   
   f. Mettre à jour v en v_new.
   
4. Calculer la valeur singulière σ en tant que norme de A @ v :
   σ = ||A @ v||

5. Retournez u, σ et v.
